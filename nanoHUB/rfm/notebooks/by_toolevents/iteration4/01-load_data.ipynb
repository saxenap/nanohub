```python
from pathlib import Path
import os
import pandas as pd
import numpy as np
from nanoHUB.rfm.functions import CacheDirs
```

```python
cached_dirs = CacheDirs()
data_dir = Path(cached_dirs.cache_dir, 'rfm/by_toolevents/iteration4')
```

```python
USERS_TO_REMOVE = ['mmc', 'clarksm', 'gridstat', 'instanton']

df = pd.read_csv(
    Path(cached_dirs.cache_dir, 'toolevents_rfm.csv'),
    index_col=False
)
display(df)
```

```python
self_study_users = pd.read_csv(Path(cached_dirs.cache_dir, "self_study_usernames.csv"))

self_study_users = self_study_users[self_study_users['username'].notna()]
self_study_users = self_study_users[~self_study_users['username'].str.startswith('-')]
```

```python
df = df[df['user'].isin(self_study_users['username'])].reset_index()
display(df)
```

```python
dfc = df[(df.first_start_date >= '2005-11-12')]
display(len(dfc))
dfc = dfc[(dfc.toolevents__lifetime > 1)]
display(len(dfc))
dfc = dfc[(dfc.job_event_days > 1)]
display(len(dfc))
dfc = dfc[~dfc.user.isna()]
dfc = dfc[~dfc.user.isin(USERS_TO_REMOVE)]
```

```python
now = pd.to_datetime('now')
last_12months = now - pd.DateOffset(months=12)

dfc['start_datetime'] = pd.to_datetime(dfc['first_start_date'])
dfc = dfc.query("(@now >= start_datetime >= @last_12months)")
display(len(dfc))
display(dfc)
```

```python
dfc.to_csv(
    Path(data_dir, "unclassified_user_data.csv"),
    index=False,
    index_label=False
)
display(pd.read_csv(
    Path(data_dir, "unclassified_user_data.csv")
))
```
