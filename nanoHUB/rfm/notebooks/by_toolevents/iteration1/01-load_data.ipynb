```python
from pathlib import Path
import os
import pandas as pd
import numpy as np
```

```python
CACHE_DIR = Path(Path(os.getenv('APP_DIR')), '.cache')
NANOHUB_CACHE = Path(CACHE_DIR, 'nanohub')
NANOHUB_METRICS_CACHE = Path(CACHE_DIR, 'nanohub_metrics')

USERS_TO_REMOVE = ['mmc', 'clarksm', 'gridstat', 'instanton']

df = pd.read_csv(
    Path(CACHE_DIR, 'toolevents_rfm.csv'),
    index_col=False
)
display(df)
```

```python
self_study_users = pd.read_csv(Path(CACHE_DIR, "self_study_usernames.csv"))

self_study_users = self_study_users[self_study_users['username'].notna()]
self_study_users = self_study_users[~self_study_users['username'].str.startswith('-')]
```

```python
df = df[df['user'].isin(self_study_users['username'])].reset_index()
display(df)
```

```python
dfc = df[(df.first_start_date >= '2005-11-12')]
display(len(dfc))
dfc = dfc[(dfc.toolevents__lifetime > 1)]
display(len(dfc))
dfc = dfc[(dfc.job_count > 1)]
display(len(dfc))
dfc = dfc[(dfc.job_event_days > 1)]
display(len(dfc))
dfc = dfc[~dfc.user.isna()]
dfc = dfc[~dfc.user.isin(USERS_TO_REMOVE)]
```

```python
dfc.to_csv(
    Path(CACHE_DIR, "unclassified_user_data.csv"),
    index=False,
    index_label=False
)
display(pd.read_csv(Path(CACHE_DIR, "unclassified_user_data.csv")))
```
