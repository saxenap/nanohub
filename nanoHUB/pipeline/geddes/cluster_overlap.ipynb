```python
import datetime

from nanoHUB.application import Application
from nanoHUB.pipeline.geddes.data import get_default_s3_client
from nanoHUB.configuration import ClusteringConfiguration
from nanoHUB.pandas_helper import remove_display_restrictions, compare_dfs
from nanoHUB.clustering.overlap import get_cluster_overlap
from nanoHUB.clustering.infra import create_clusters_repository
import pandas as pd


remove_display_restrictions(pd)

application = Application.get_instance()
s3_client = get_default_s3_client(application)

config = ClusteringConfiguration()

bucket_name_raw = config.bucket_name_raw
bucket_name_processed = config.bucket_name_processed


alg_types = ['mike', 'xufeng']
```
