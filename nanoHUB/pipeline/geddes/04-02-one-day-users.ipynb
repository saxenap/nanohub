```python
import pandas as pd
import numpy as np
from pprint import pprint
from nanoHUB.application import Application
from nanoHUB.configuration import ClusteringConfiguration
from nanoHUB.pipeline.geddes.data import get_default_s3_client
from nanoHUB.dataaccess.lake import S3FileMapper
from nanoHUB.clustering.infra import create_clusters_repository, add_cluster_info
from ast import literal_eval

pd.set_option('display.max_columns', None)

# logger.debug('Testing')
application = Application.get_instance()
nanohub_db = application.new_db_engine('nanohub')

#salesforce = application.new_salesforce_engine()

s3_client = get_default_s3_client(application)
raw_mapper = S3FileMapper(s3_client, ClusteringConfiguration().bucket_name_raw)
processed_mapper = S3FileMapper(s3_client, ClusteringConfiguration().bucket_name_processed)

cluster_repo = create_clusters_repository(application, ClusteringConfiguration().bucket_name_processed)

derived_data_file_path = 'derived_data_for_users.csv'
```

```python
all_users_df = processed_mapper.read(derived_data_file_path,converters={"clusters": literal_eval})

display("Number of all users: %d" % len(all_users_df))
display(all_users_df.head())
```

```python
sql_query = '''
SELECT *
FROM (
       SELECT DISTINCT one_day_users.user,
                       one_day_users.first_session_datetime,
                       one_day_users.last_session_datetime
       FROM nanohub_metrics.toolstart AS all_users
              JOIN (
                SELECT DISTINCT user,
                       MAX(datetime) AS last_session_datetime,
                       MIN(datetime) AS first_session_datetime
                FROM nanohub_metrics.toolstart
                GROUP BY user
         ) one_day_users
                   ON one_day_users.user = all_users.user
                        AND one_day_users.last_session_datetime <= one_day_users.first_session_datetime + INTERVAL 1 DAY
       WHERE all_users.datetime IS NOT NULL
         AND all_users.datetime != '0000-00-00 00:00:00'
         AND all_users.datetime >= '2020-04-01 00:00:00'
         AND all_users.datetime <= '2021-03-31 00:00:00'
     ) final
;
'''
```

```python
one_day_users_df = pd.read_sql_query(sql_query, nanohub_db)

display(len(one_day_users_df))
display(one_day_users_df.head())
display(one_day_users_df.tail())
```

```python
sql_query = '''
SELECT *
FROM (
       SELECT DISTINCT one_day_users.user,
                       one_day_users.first_session_datetime,
                       one_day_users.last_session_datetime
       FROM nanohub_metrics.toolstart AS all_users
              JOIN (
         SELECT DISTINCT user,
                         MAX(datetime) AS last_session_datetime,
                         MIN(datetime) AS first_session_datetime
         FROM nanohub_metrics.toolstart
         WHERE datetime IS NOT NULL
           AND datetime != '0000-00-00 00:00:00'
         GROUP BY user
         HAVING MAX(datetime) <= (MIN(datetime) + INTERVAL 1 DAY)
       ) one_day_users
                   ON one_day_users.user = all_users.user
       WHERE all_users.datetime IS NOT NULL
         AND all_users.datetime != '0000-00-00 00:00:00'
         AND all_users.datetime >= '2020-04-01 00:00:00'
         AND all_users.datetime <= '2021-03-31 00:00:00'
     ) final
;
'''
```

```python
one_day_users_df = pd.read_sql_query(sql_query, nanohub_db)

display(len(one_day_users_df))
display(one_day_users_df.head())
display(one_day_users_df.tail())
```

```python
sql_query = '''
SELECT toolstart.user,
       toolstart.first_session_datetime,
       toolstart.last_session_datetime
FROM
     (
      SELECT DISTINCT user,
                      MAX(datetime) AS last_session_datetime,
                      MIN(datetime) AS first_session_datetime
      FROM nanohub_metrics.toolstart
      WHERE datetime IS NOT NULL
        AND datetime != '0000-00-00 00:00:00'
      GROUP BY user
      HAVING MAX(datetime) <= (MIN(datetime) + INTERVAL 1 DAY)
      ) AS toolstart
WHERE toolstart.first_session_datetime >= '2020-04-01 00:00:00'
  AND toolstart.last_session_datetime <= '2021-03-31 00:00:00'
;
'''
```

```python
one_day_users_df = pd.read_sql_query(sql_query, nanohub_db)

display(len(one_day_users_df))
display(one_day_users_df.head())
display(one_day_users_df.tail())
```
