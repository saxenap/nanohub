```python
import pandas as pd
import datetime
from nanoHUB.application import Application

application = Application.get_instance()
nanohub_db = application.new_db_engine('nanohub')
```

```python
query = '''
SELECT users.id, users.username, users.email, users.registerDate, users.lastvisitDate, users.activation, users.approved, users.block
FROM nanohub.jos_users users
WHERE users.id NOT IN (SELECT user_id FROM nanohub.jos_reply_access_codes WHERE page_id = 2)
'''

users_df = pd.read_sql_query(query, nanohub_db)
display('Total number of missing users with reply_access_codes: %s' % len(users_df))
display(users_df.head(3))
display(users_df.tail(3))
```

```python
query = '''
SELECT users.id, users.username, users.email, users.registerDate, users.lastvisitDate, users.activation, users.approved, users.block
FROM nanohub.jos_users users
WHERE users.id NOT IN (SELECT user_id FROM nanohub.jos_reply_access_codes WHERE page_id = 2)
    AND users.registerDate >= '2022-05-01 00:00:00'
'''

users_df = pd.read_sql_query(query, nanohub_db)
display('Number of missing users with reply_access_codes: %s' % len(users_df))
display(users_df.head(3))
display(users_df.tail(3))
```

```python
registered_since = '2022-05-01 00:00:00'
query = '''
SELECT users.id, users.username, users.email, users.registerDate, users.lastvisitDate, users.activation, users.approved, users.block
FROM nanohub.jos_users users
WHERE users.registerDate >= '%s'
'''

users_df = pd.read_sql_query(query % registered_since, nanohub_db)
display('Number of registered users since %s: %s' % (registered_since, len(users_df)))
display(users_df.head(3))
display(users_df.tail(3))
```
