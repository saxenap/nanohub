```python
from pathlib import Path
import pandas as pd
import os 
import ast
import itertools


data_dir = Path(os.getenv('APP_DIR') + '/nanoHUB/data/clusters')
overlap_dir = data_dir / 'mike-xufeng-overlap' / 'by_semester' 
```

```python

overlap_df = pd.read_csv(overlap_dir / '2010-01-01_2010-07-01.csv')
display(overlap_df.head(2))
```

```python
columns = ['CombinedMembers']

df = overlap_df
df[columns] = df[columns].applymap(lambda x: ast.literal_eval(x))
display(df.head(2))
```

```python
combined_members = df.groupby('MClusterID')['CombinedMembers'].agg(list).reset_index()
display(combined_members.describe())
```

```python
combined_members.apply(lambda a: a[:])
```

```python
display(combined_members)
```

```python
combined_members['CombinedMembers'] = combined_members['CombinedMembers'].apply(lambda x: list(itertools.chain.from_iterable(x)))                      
display(combined_members)
```
